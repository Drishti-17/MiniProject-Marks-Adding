```python
import csv

csv_file_path = '/content/data.csv'  

data_dict = {}

with open(csv_file_path, 'r') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    for row in csv_reader:
        roll_number = row['Roll Num']
        marks = int(row['Marks'])

        if roll_number in data_dict:
            data_dict[roll_number] += marks
        else:
            data_dict[roll_number] = marks

# Print the dictionary
for roll_number, total_marks in data_dict.items():
    print(f'Roll Number: {roll_number}, Total Marks: {total_marks}')

```
